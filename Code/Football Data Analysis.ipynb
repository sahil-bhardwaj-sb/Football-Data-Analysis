### Insight:

```python
#Total Goals/assists per MATCH is/are NOT the most accurate metric for any player's performance.
#In Football, it is possible that a player played for 90 minutes in one match and played for only 5 minutes (as a substitute)
#in the next match. Hence, his NUMBER OF MATCHES will be = 2, but his NUMBER OF MINUTES played will be = 95.

#Let us consider he scored 1 goal during the interval mentioned. So now, if we calculate his Goal/Match metric, it will be:
# 0.5 goal per match. However, if we calculate his Goal/90 Minutes metric, it will be:
# (1/95)*90 = 0.95 goal per 90 minutes.
```

## Adding a function to increase versatility

```python
#This function plots the graph for only those players you send as a parameter in a list to it.
def PlayerMetric(playerList):
    for name in playerList:
        plt.plot(Goals[Pdict[name]], c='Black', ls = '--', marker = 's', ms = 8, label = name) 
    plt.legend(loc = 'upper left', bbox_to_anchor = (1, 1))    
    plt.xticks(list(range(0,6)),Seasons, rotation = 'vertical')
    plt.show()
```

```python
PlayerMetric(['HarryKane', "Neymar"])
```

```python
#As it can be seen, the colours and markers need to be separated:
```

```python
def PlayerMetric(playerList):
    #Adding a Colours dictionary (for Colours) dictionary to specify the colour for each player.
    Colours = {"LionelMessi":"Black","CristianoRonaldo":"Blue","RobertL":"Green",
               "HarryKane":"Red","Neymar":"Yellow"}
    
    #Adding a Marks dictionary (for Markers) to specify the marker for each player.
    Marks = {"LionelMessi":"s","CristianoRonaldo":"o","RobertL":"*",
               "HarryKane":"^","Neymar":"D"}
    
    for name in playerList:                                      #Iterating over each player.
        plt.plot(Goals[Pdict[name]], c=Colours[name], ls = '--', marker = Marks[name], ms = 8, label = name)    
    plt.legend(loc = 'upper left', bbox_to_anchor = (1, 1))    
    plt.xticks(list(range(0,6)),Seasons, rotation = 'vertical')
    plt.show() 
```

```python
PlayerMetric(['HarryKane', "Neymar", "LionelMessi"])
```

### Fixing the inputs:

```python
#The visualization will be made eben more versatile such that you could pass 'data' = goals, assists, yellow cards etc 
#and the visualization will adjust accordingly to show the desired metric.
```

```python
def PlayerMetric(data, playerList):                   #Creating a new parameter 'data' for the metric to be specified. 
    Colours = {"LionelMessi":"Black","CristianoRonaldo":"Blue","RobertL":"Green",
               "HarryKane":"Red","Neymar":"Yellow"}

    Marks = {"LionelMessi":"s","CristianoRonaldo":"o","RobertL":"*",
               "HarryKane":"^","Neymar":"D"}
    
    for name in playerList:
        plt.plot(data[Pdict[name]], c=Colours[name], ls = '--', marker = Marks[name], ms = 8, label = name)    
    plt.legend(loc = 'upper left', bbox_to_anchor = (1, 1))    
    plt.xticks(list(range(0,6)),Seasons, rotation = 'vertical')
    plt.show() 
```

```python
PlayerMetric(YellowCards, ['HarryKane', "Neymar", "LionelMessi"])
```

```python
#Adding a default paramter to show the desired metric for every player if nothing is passed inside the playerList.
```

```python
def PlayerMetric(data, playerList = Players):          #Passing the Players list as a default value to the playerList parameter.

    Colours = {"LionelMessi":"Black","CristianoRonaldo":"Blue","RobertL":"Green",
               "HarryKane":"Red","Neymar":"Yellow"}

    Marks = {"LionelMessi":"s","CristianoRonaldo":"o","RobertL":"*",
               "HarryKane":"^","Neymar":"D"}
    
    for name in playerList:
        plt.plot(data[Pdict[name]], c=Colours[name], ls = '--', marker = Marks[name], ms = 8, label = name)    
    plt.legend(loc = 'upper left', bbox_to_anchor = (1, 1))    
    plt.xticks(list(range(0,6)),Seasons, rotation = 'vertical')
    plt.show() 
```
